```python
get_online_metrics = False
session_idx = 1
path = '../../Datasets/Pilots/Pilot_2/Session_{}/formatted_filt_250Hz/'.format(session_idx) 
# {formatted_filt_250Hz, action_vs_rest_filt_250Hz}

train_score = { 'accuracy' : [], 'kappa' : [] }
test_score = { 'accuracy' : [], 'kappa' : [] }
online_scores = { 'accuracy' : [], 'kappa' : [] }
conf_matrices = {}
train_times = []
test_times = []

# Parameters
fs = 250
start=2.
end=6.
n_crops = 1
crop_len = 1.
online_stride = fs//10
batch_size = 32

# Load dataset (full trial and extracted MI ROI)
X_full, y, _, _, _, _ = EEGDataset(1, data_path=path, rereferencing=True, 
                                   standardization=True, filt=False,
                                   load_test=False).load_dataset()

X = X_full[:, :, int(start*fs): int(end*fs)]
n_classes = len(np.unique(y))

# Get more robust validation metrics via k-folding (no pre-defined test set like in BCIC)
skf = StratifiedKFold(n_splits=5, shuffle=True)
idx_kfold = 0
for idx_train, idx_test in skf.split(X, y):
    # Get k-fold data (with cropping)
    print('Fold ', idx_kfold+1)
    idx_kfold += 1
    X_train, y_train = X[idx_train], y[idx_train]
    X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, 
                                                          random_state=0, shuffle=False, stratify=y_train)
    X_test, y_test = X[idx_test], y[idx_test]
    X_online, y_online = X_full[idx_test], y[idx_test]
    
    # Cropping the data
    X_train_crops, y_train_crops, trial2crops_train = cropping(X_train, y_train, fs, n_crops,
                                                               crop_len, shuffled=True)
    X_valid_crops, y_valid_crops, trial2crops_valid = cropping(X_valid, y_valid, fs, n_crops,
                                                               crop_len, shuffled=False)
    X_test_crops, y_test_crops, trial2crops_test = cropping(X_test, y_test, fs, n_crops,
                                                            crop_len, shuffled=False)

    # Reshape in Keras format (n_trials, n_filter, n_channels, n_samples)
    n_trials, n_channels, n_samples = X_train_crops.shape
    X_train_crops = np.reshape(X_train_crops, (n_trials, 1, n_channels, n_samples))
    print(X_train_crops.shape)

    n_trials, n_channels, n_samples = X_valid_crops.shape
    X_valid_crops = np.reshape(X_valid_crops, (n_trials, 1, n_channels, n_samples))
    print(X_valid_crops.shape)
    
    n_trials, n_channels, n_samples = X_test_crops.shape
    X_test_crops = np.reshape(X_test_crops, (n_trials, 1, n_channels, n_samples))
    print(X_test_crops.shape)

    # Load architecture
    model = ShallowConvNet(nb_classes=n_classes, n_channels=n_channels, n_samples=n_samples)

    # Training parameters
    es = EarlyStopping(patience=50, monitor='val_loss', verbose=1)
    mc = ModelCheckpoint('best_model.h5', 'val_loss', verbose=1, save_best_only=True, save_weights_only=True)
    cam = CropsAveragingMetrics(validation_data=(X_valid_crops, y_valid),
                                trial2crops=trial2crops_valid, n_trials=len(y_valid))
    
    model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.Adam(), metrics=['accuracy'])
    
    print('Fitting model...')
    train_time = time.time()
    train_gen = datagenerator(X_train, y_train, fs, n_crops, crop_len, batch_size)
    valid_gen = datagenerator(X_valid, y_valid, fs, n_crops, crop_len, batch_size)
    hist = model.fit_generator(generator=train_gen, validation_data=valid_gen, callbacks=[es,mc], 
                               steps_per_epoch=len(X_train)/batch_size, validation_steps=len(X_valid)/batch_size)
                               #use_multiprocessing=True,
                               #workers=6)
    
    #hist = model.fit(x=X_train_crops, y=to_categorical(y_train_crops), batch_size=32, epochs=5, verbose=0, 
    #                 validation_data=(X_valid_crops, to_categorical(y_valid_crops)), callbacks=[es, cam])
    train_time = time.time() - train_time
    train_times.append(1e3*train_time/X_train_crops.shape[0])
    print('Took {:.1f}s to fit the model (about {:.1f}ms per trial or {:.1f}ms per crop).\n'.format(train_time, 
                                                                                            1e3*train_time/X_train.shape[0],
                                                                                            1e3*train_time/X_train_crops.shape[0]))

    best_epoch = np.argmin(hist.history['val_loss'])
    print('Best epoch: {} - Valid loss: {:.2} - Valid acc: {:.2}'.format(best_epoch, 
                                                                        hist.history['val_loss'][best_epoch],
                                                                        hist.history['val_acc'][best_epoch]))

    # Load best model
    model.load_weights('best_model.h5')

    # Validation (train set metrics) 
    y_pred_crops = np.argmax(model.predict(X_train_crops), axis=1)
    y_pred_train = np.empty_like(y_train)
    for trial_idx in range(len(y_train)):
        labels, cnts = np.unique(y_pred_crops[trial2crops_train[trial_idx]], return_counts=True)
        y_pred_train[trial_idx] = labels[np.argmax(cnts)]

    # Validation (test set metrics with prediction averaging)
    test_time = time.time()
    y_pred_crops = np.argmax(model.predict(X_test_crops), axis=1)
    y_pred_test = np.empty_like(y_test)
    for trial_idx in range(len(y_test)):
        labels, cnts = np.unique(y_pred_crops[trial2crops_test[trial_idx]], return_counts=True)
        y_pred_test[trial_idx] = labels[np.argmax(cnts)]
    test_time = time.time() - test_time
    test_times.append(1e3*test_time/X_test_crops.shape[0])
    print('Took {:.1f}ms to perform inference (about {:.1f}ms per trial or {:.1f}ms per crop).'.format(1e3*test_time, 
                                                                                1e3*test_time/X_test.shape[0],
                                                                                1e3*test_time/X_test_crops.shape[0]))

    # Saving metrics and confusion matrices for each k-fold
    train_score['accuracy'].append(accuracy_score(y_train, y_pred_train))
    train_score['kappa'].append(cohen_kappa_score(y_train, y_pred_train))
    test_score['accuracy'].append(accuracy_score(y_test, y_pred_test))
    test_score['kappa'].append(cohen_kappa_score(y_test, y_pred_test))
    conf_matrices['y_pred{}'.format(idx_kfold)] = y_pred_test
    conf_matrices['y_true{}'.format(idx_kfold)] = y_test

    # Getting online scores (temporal accuracy averaged over trials)
    if get_online_metrics:
        scores = compute_online_metrics(model, X_online, y_online, int(crop_len*fs), online_stride)
        if len(online_scores['accuracy'])==0:
            online_scores = scores
        else:
            online_scores['accuracy'] = np.sum([online_scores['accuracy'], scores['accuracy']], axis=0)
            online_scores['kappa'] = np.sum([online_scores['kappa'], scores['kappa']], axis=0)

    # Logging for current Kfold
    print('Pilot 1 - Train: {:.3f}/{:.3f} - Test: {:.3f}/{:.3f}. \n'.format(train_score['accuracy'][-1],
                                                                            train_score['kappa'][-1],
                                                                            test_score['accuracy'][-1],
                                                                            test_score['kappa'][-1]))

print('Mean training score: {:.3f} (accuracy) - {:.3f} (kappa)'.format(np.mean(train_score['accuracy']),
                                                                    np.mean(train_score['kappa'])))
print('Mean test score: {:.3f}+-{:.3f} (accuracy) - {:.3f}+-{:.3f} (kappa)'.format(np.mean(test_score['accuracy']),
                                                                                np.std(test_score['accuracy']),
                                                                                np.mean(test_score['kappa']),
                                                                                np.std(test_score['kappa'])))
print('Mean training time per trial: {:.1f}ms'.format(np.mean(train_times)))
print('Mean testing time per trial: {:.1f}ms'.format(np.mean(test_times)))
```
